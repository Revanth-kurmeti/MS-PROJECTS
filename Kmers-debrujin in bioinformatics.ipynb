# 2.2 Generating K-mers

```python
# Open the input file and read the contents into a list of reads
with open("C:/Users/revan/Downloads/1002068038.txt", "r") as file:
    reads = file.read().splitlines()

# Initialize a set to store the unique k-mers
kmers = list()

# Iterate through each read and extract all k-mers of length 2
for read in reads:
    for i in range(len(read)-1):
        kmer = read[i:i+2]
        kmers.append(kmer)

# Write the k-mers to a file called "kmers.txt"
with open("kmers.txt", "w") as file:
    for kmer in kmers:
        file.write(kmer + "\n")
```

# 2.3 Generating a De Bruijn Graph 

```python
import networkx as nx
import matplotlib.pyplot as plt

# Read k-mers from file
with open("C:/Users/revan/1002068038.txt", "r") as file:
    kmers = file.read().splitlines()

# Generate De Bruijn graph
graph = nx.DiGraph()
for kmer in kmers:
    prefix = kmer[:-1]
    suffix = kmer[1:]
    graph.add_edge(prefix, suffix)

# Plot De Bruijn graph
pos = nx.circular_layout(graph)
nx.draw_networkx_nodes(graph, pos, node_size=500)
nx.draw_networkx_edges(graph, pos, width=2)
nx.draw_networkx_labels(graph, pos, font_size=16, font_family='sans-serif')
plt.axis('off')
plt.show()
```
